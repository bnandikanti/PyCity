# PyCity Schools Analysis

* As a whole, schools with higher budgets, did not yield better test results. By contrast, schools with higher spending per student actually (\$645-675) underperformed compared to schools with smaller budgets (<\$585 per student).

* As a whole, smaller and medium sized schools dramatically out-performed large sized schools on passing math performances (89-91% passing vs 67%).

* As a whole, charter schools out-performed the public district schools across all metrics. However, more analysis will be required to glean if the effect is due to school practices or the fact that charter schools tend to serve smaller student populations per school.

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
school_data_complete.columns
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_schools = school_data["school_name"].value_counts().count()
total_students = student_data["Student ID"].value_counts().count()
total_budget = school_data["budget"].sum()
average_math_score = student_data["math_score"].mean()
average_reading_score = student_data["reading_score"].mean()
overall_passing_rate = (average_math_score + average_reading_score) / 2
percent_math_pass = (student_data.loc[(student_data["math_score"] >= 70), "math_score"].count() / total_students) * 100
percent_reading_pass = (student_data.loc[(student_data["reading_score"] >= 70), "reading_score"].count() / total_students) * 100

district_summary = pd.DataFrame([{"Total Schools" : total_schools , "Total Students": total_students, "Total Budget": total_budget,
                     "Average Math Score" : average_math_score, "Average Reading Score" : average_reading_score, 
                     "Overall Passing Rate": overall_passing_rate, "% Passing Math" : percent_math_pass,
                     "% Passing Reading" : percent_reading_pass}])

district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["% Passing Math"] = district_summary["% Passing Math"].map("{:.2f}".format)
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].map("{:.2f}".format)
district_summary["Overall Passing Rate"] = district_summary["Overall Passing Rate"].map("{:.2f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:.2f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:.2f}".format)
district_summary
```

## Math Scores by Grade

```python
math_series_9th = school_data_complete[school_data_complete['grade'] == '9th']
math_9th_gby = math_series_9th.groupby("school_name")
math_by_9th = math_9th_gby.mean()
math_series_10th = school_data_complete[school_data_complete['grade'] == '10th']
math_10th_gby = math_series_10th.groupby("school_name")
math_by_10th = math_10th_gby.mean()
math_series_11th = school_data_complete[school_data_complete['grade'] == '11th']
math_11th_gby = math_series_11th.groupby("school_name")
math_by_11th = math_11th_gby.mean()
math_series_12th = school_data_complete[school_data_complete['grade'] == '12th']
math_12th_gby = math_series_12th.groupby("school_name")
math_by_12th = math_gby.mean()
math_clean_df = pd.DataFrame({"9th" : math_by_9th["math_score"], "10th" : math_by_10th["math_score"],
                   "11th" : math_by_11th["math_score"], "12th" : math_by_12th["math_score"]})
math_clean_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
school_summary2 = school_data.copy()
students = school_data_complete["school_name"].value_counts()
school_budget = school_data_complete.groupby(["school_name"]).budget.mean()
student_budget = school_budget / students
df = pd.DataFrame()
df["student_budget"] = student_budget

total_students = school_data_complete["school_name"].value_counts()
average_math_score = school_data_complete.groupby(["school_name"]).math_score.mean()
average_reading_score = school_data_complete.groupby(["school_name"]).reading_score.mean()
df["Average Math Score"] = average_math_score
df["Average Reading Score"] = average_reading_score
math_pass = school_data_complete[school_data_complete["math_score"]  >= 70 ]
reading_pass = school_data_complete[school_data_complete["reading_score"]  >= 70 ]
math_gby= math_pass.groupby("school_name")
reading_gby= reading_pass.groupby("school_name")
percent_math_pass = (math_gby["math_score"].count() / total_students) * 100
percent_reading_pass = (reading_gby["reading_score"].count() / total_students) * 100
overall_percent_pass =  (percent_math_pass + percent_reading_pass) /2

df["% Passing Math"] = percent_math_pass
df["% Passing Reading"] =  percent_reading_pass
df["Overall Passing Rate"] = overall_percent_pass

df["Spending Ranges {Per Student}"] = pd.cut(df["student_budget"], spending_bins, labels=group_names)
spending_summary = df.groupby("Spending Ranges {Per Student}")
spending_summary.min()

```
